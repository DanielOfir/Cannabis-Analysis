```python
import scrapy
import json
import time
from scrapy.crawler import CrawlerProcess
from scrapy.selector import Selector
from selenium import webdriver
```

```python
class StrainsSpider(scrapy.Spider):

    name = "strains"

    with open('strain_list.txt','r') as f:
        strains = json.loads(f.read())

    start_urls = ['https://www.leafly.com' + x for x in strains[:5]]


    def __init__(self):
        # self.browser = webdriver.Chrome()
        self.browser = webdriver.Chrome('/Users/daniel.ofir/.wdm/drivers/chromedriver/mac64/90.0.4430.24/chromedriver')

    
    def parse(self,response):

        # page = response.url.split('/')[-1]
        self.browser.get(response.url)
        time.sleep(2)
        # time.sleep(3)
        filename = f'{response.url}.html'
        # page = self.browser.page_source
        with open(filename,'wb') as f:
            # f.write(response.body)
            f.write(response.body)

        self.browser.close()
    
```

```python
# process = CrawlerProcess({
#     'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)'
#     })
process = CrawlerProcess()
process.crawl(StrainsSpider)
process.start()
```
